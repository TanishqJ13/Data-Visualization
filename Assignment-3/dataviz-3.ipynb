```python
import pandas as pd
import numpy as np

import plotly.graph_objects as go
import plotly.express as px

import networkx as nx
from scipy.io import mmread

import matplotlib.pyplot as plt
from matplotlib import cm

import seaborn as sns
from tqdm import tqdm
import random

import ortools
from ortools.constraint_solver import pywrapcp, routing_enums_pb2
from scipy.spatial.distance import squareform, pdist
```

```python
mat = mmread('../input/dataset/bio-diseasome.mtx')
```

```python
G = nx.from_numpy_matrix(np.array(mat.todense()))
plt.figure(figsize=(30,30))
cmap = plt.cm.coolwarm
colors = [n for n in range(len(G.nodes()))]
pos = nx.spring_layout(G)
nx.draw_networkx(G,pos, node_size=len(G.nodes()), cmap = cmap, node_color=colors, edge_color='grey', font_size=5, width=2, alpha=1)
plt.show()
plt.savefig('./output8.png')
```

```python
plt.figure(figsize=(30,30))
cmap = plt.cm.coolwarm
colors = [n for n in range(len(G.nodes()))]
pos = nx.circular_layout(G)
nx.draw_networkx(G,pos, node_size=len(G.nodes()), cmap = cmap, node_color=colors, edge_color='grey', font_size=5, width=2, alpha=1)
plt.show()
plt.savefig('./output7.png')
```
